```python
import pennylane as qml
from pennylane import numpy as np
```

```python
def F_kn(q0, q1, k, n):
    qml.PhaseShift(-2j*np.pi*k/n, wires = q1)
    qml.Hadamard(wires = q0)
    qml.Hadamard(wires = q1)
    qml.CNOT(wires = [q0, q1])
    qml.Hadamard(wires = q0)
    qml.Hadamard(wires = q1)
    qml.ctrl(qml.Hadamard, control = q0)(q1)
    qml.Hadamard(wires = q0)
    qml.Hadamard(wires = q1)
    qml.CNOT(wires = [q0, q1])
    qml.CZ(wires = [q0, q1])
    
```

```python
def Fdag(wires, k): # for n = 4
    q0 = wires[0]
    q1 = wires[1]
    
    #these two give F2 which is hermitian so F_2^dag = F_2
    #FIGURE OUT HOW THIS RELATES TO QFT BETTER
    qml.FermionicSWAP(phi = np.pi, wires = wires) 
    qml.SingleExcitation(phi = np.pi/2, wires = wires)

    #Twiddle factor for the odd sector (for k = 0, its just identity so ignore)
    #Take the dag because this one is not hermitian
    if k == 1: #for n = 4 specifically. Not general to all n       
        qml.adjoint(qml.S(wires = q0)) #pi/2

dev = qml.device("default.qubit", wires=2)
@qml.qnode(dev)
def circuit(k):
    Fdag([0, 1], k)
    return qml.state()

# Extract matrix
Uold = qml.matrix(circuit)(k = 0)
Uold
```

```python


dev = qml.device("default.qubit", wires=2)

@qml.qnode(dev)
def circuit(k, n):
    F_kn(0, 1, k, n)
    return qml.state()

# Extract matrix
U = qml.matrix(circuit)(k = 0, n = 4)

U
```

```python
np.allclose(U, Uold)
```
