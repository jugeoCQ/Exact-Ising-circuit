```python
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt 

np.set_printoptions(precision = 3, suppress = True)
```

```python
def F_kn(q0, q1, k, n, dag = False):#figure 3 of the XY paper. Different convention than fig.8 of the Ising paper? 
    if k!= 0 and dag == False:
        qml.PhaseShift(-2*np.pi*k/n, wires = q1) #This is an S gate for k = 1 n = 2
    if k!=0 and dag == True:
        qml.PhaseShift(+2*np.pi*k/n, wires = q1) #dagger
        
    
    qml.Hadamard(wires = q0)
    qml.Hadamard(wires = q1)
    qml.CNOT(wires = [q0, q1])
    qml.Hadamard(wires = q0)
    qml.Hadamard(wires = q1)
    
    qml.ctrl(qml.Hadamard, control = q0)(q1)
    
    qml.Hadamard(wires = q0)
    qml.Hadamard(wires = q1)
    qml.CNOT(wires = [q0, q1])
    qml.Hadamard(wires = q0)
    qml.Hadamard(wires = q1)
    
    qml.CZ(wires = [q0, q1])

dev = qml.device("default.qubit", wires = 2)
@qml.qnode(dev)
def circuit(k, n, dag):
    F_kn(0, 1, k, n, dag = dag)
    return qml.state()

# Double check matrix matches circuit
U = qml.matrix(circuit)(k = 1, n = 4, dag = False)
Udag = qml.matrix(circuit)(k = 1, n = 4, dag = True)
print(U, '\n \n', Udag)
```

```python
def fSWAPSFT(numq):
    #create a circuit with numq qubits
    #create a count0 which start at 1, the fSWAP always start at q1
    count0=1
    #the number of fSWAPS in the first layer is equal to numq/2 -1 and then it decreases by one until there is only
    #1 fSWAP.
    for j in range(int(numq/2.0)-1,0,-1):
        #count1, the first fSWAP is added in the qbit(count0) 
        count1=count0
        for i in range(1,j+1):
            #add the fSWAP in the circuit
            print("fswap", (count1,count1+1))
            #fSWAP gate is a 2qbit gate, the next gate must be added 
            count1+=2
        #After the first layer the initial qbit is the ne one
        count0+=1
    return None
```

```python
def downfallFSWAP(nq,qbinit,numFSWAP):
    #initialize the quantum circuit with nq qubits
    #save as variable where the circuit starts
    qb0=qbinit
    #loop to add all gates
    for i in range(0,numFSWAP):
        #add the FSWAP gate in the qubits qb0 and qb0+1
        print("FSWAP"(qb0,qb0+1))
        #sum 1 to qb0
        qb0+=1
    return None

downfallFSWAP(4, 6)
```
