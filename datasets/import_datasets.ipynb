```python
import pennylane as qml
from pennylane import numpy as np
```

```python
def extract_data(chain_size):
    if chain_size == 4:
        layout = "1x4"
    elif chain_size == 8:
        layout = "1x8"
    elif chain_size == 16:
        layout = "1x16"
    
    ising_datasets= qml.data.load("qspin", sysname = "Ising", periodicity = "closed", lattice = "chain", layout = layout)

    return ising_datasets[0]
```

```python
chain_sizes = [4, 8, 16]
ground_state_files  = ['gs4_closed', 'gs8_closed', 'gs16_closed']
gs_energy_files  = ['gsenergy4_closed', 'gsenergy8_closed', 'gsenergy16_closed']

for i in range(len(chain_sizes)):
    data = extract_data(chain_sizes[i])
    gs = data.ground_states
    gs_energy = data.ground_energies

    np.save(ground_state_files[i], gs)
    np.save(gs_energy_files[i], gs_energy)
    
```
